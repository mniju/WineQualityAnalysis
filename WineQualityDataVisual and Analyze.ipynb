### Visualizing the Wine Quality Dataset

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

Read the Data

```python
# read in the raw data
wine_data = pd.read_csv("winequality/winequality-white.csv", encoding='latin-1')
df = pd.read_csv('winequality/winequality-white.csv',sep=';',quotechar='"')

# print the head
print(df.head()) 
```

```python
import seaborn as sns
sns.distplot(df['quality'])
```

```python
sns.set(style="darkgrid")
sns.countplot(df['quality'])
```

```python
sns.swarmplot(x='quality', y='alcohol', data=df)
```

```python
sns.swarmplot(x='quality', y='total sulfur dioxide', data=df)
```

```python
sns.swarmplot(x='total sulfur dioxide',y='quality',data=df)
```

```python
g = sns.factorplot(x='quality', y='total sulfur dioxide', data=df)
```

```python
g = sns.factorplot(x='quality', y='alcohol', data=df)
```

```python
g = sns.factorplot(x='alcohol', y='total sulfur dioxide',hue = 'quality', data=df)
```
